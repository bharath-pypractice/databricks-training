```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Sravan", 25, "Hyderabad", "Data Engineer", 55000, "2023-01-15", "IT"),
    (2, "Ravi", 28, "Bangalore", "Software Engineer", 72000, "2022-11-10", "IT"),
    (3, "Priya", 24, "Chennai", "Analyst", 48000, "2023-03-12", "Analytics"),
    (4, "Kiran", 30, "Pune", "Manager", 85000, "2021-09-20", "Management"),
    (5, "Sneha", 27, "Mumbai", "HR", 45000, "2020-05-18", "HR")
]

columns = [
    "emp_id",
    "emp_name",
    "age",
    "city",
    "designation",
    "salary",
    "joining_date",
    "department"
]

df = spark.createDataFrame(data, columns)

# df = df.withColumnRenamed("joining_date", "doj")

df.show()
```

```python
df = df.withColumnRenamed("city", "location")
df.show()
```

```python
df = df.withColumnRenamed("age", "employee_age")
df.show()
```

```python
df = df.withColumnRenamed("emp_name", "employee_name")

df = df.withColumnRenamed("joining_date", "doj")

df = df.withColumnRenamed("designation", "job_role")

df.show()
```

```python
df.withColumn("bonus",df.salary * 0.10).show()
```

```python
df.withColumn("annual_salary",df.salary * 12).show()
```

```python
df.withColumn("tax",df.salary * 0.05).show()
```

```python
df.withColumn("updated_salary",df.salary + 5000).show()
```

```python
df.withColumn(
    "salary_category",
    when(df.salary > 80000, "High")
    .when(df.salary >= 50000, "Medium")
    .otherwise("Low")
).show()
```

```python
df.withColumn(
    "age_group",
    when(df.age <= 18, "Young")
    .otherwise("Adult")
).show()
```

```python
df.withColumn(
    "location",
    concat(df.city,lit("-"),df.department)
).show()
```

```python
df.withColumn("salary_increment",df.salary * 0.15).show()
```

```python
df.withColumn(
    "experience_status",
    when(
        year(to_date(df.joining_date)) < 2022,
        "Experienced"
    ).otherwise("Fresher")
).show()
```

```python
df.withColumn(
    "name_length",
    length(df.emp_name)
).show()
```

```python

df.withColumn(
    "is_high_salary",
    when(df.salary > 70000, True)
    .otherwise(False)
).show()
```

```python
df.withColumn(
    "joining_year",
    year(to_date(df.joining_date))
).show()
```

```python
df.withColumn(
    "salary_after_tax",
    df.salary - (df.salary * 0.05)
).show()
```

```python
df.withColumn(
    "salary_after_tax",
    df.salary - (df.salary * 0.05)
).show()
```

```python
from pyspark.sql.functions import when

df.withColumn(
    "department_code",
    when(df.department == "IT", "IT01")
    .when(df.department == "HR", "HR01")
    .when(df.department == "Analytics", "AN01")
    .when(df.department == "Management", "MG01")
    .otherwise("OT01")
).show()
```

```python
df.withColumn(
    "double_salary",
    df.salary * 2
).show()
```
