```python
from pyspark.sql.functions import *

data = [
    (1, "Sravan", 25, "Hyderabad", "Data Engineer", 55000, "2023-01-15", "IT"),
    (2, "Ravi", 28, "Bangalore", "Software Engineer", 72000, "2022-11-10", "IT"),
    (3, "Priya", 24, "Chennai", "Analyst", 48000, "2023-03-12", "Analytics"),
    (4, "Kiran", 30, "Pune", "Manager", 85000, "2021-09-20", "Management"),
    (5, "Sneha", 27, "Mumbai", "HR", 45000, "2020-05-18", "HR")
]

columns = ["emp_id","emp_name","age","city","designation","salary","joining_date","department"]

df = spark.createDataFrame(data, columns)
```

```python
df.withColumn("salary", col("salary").cast("string")).printSchema()

df.withColumn("age", col("age").cast("double")).printSchema()

df.withColumn("joining_date", col("joining_date").cast("date")).printSchema()

df.withColumn("emp_id", col("emp_id").cast("string")).printSchema()

df.withColumn("salary", col("salary").cast("int")).printSchema()

df.withColumn("age", col("age").cast("string")).printSchema()

df.withColumn("joining_date", col("joining_date").cast("timestamp")).printSchema()

df.withColumn("salary", col("salary").cast("float")).printSchema()

df.withColumn("emp_id", col("emp_id").cast("long")).printSchema()
```

```python
df.withColumn("emp_id", col("emp_id").cast("string")) \
  .withColumn("age", col("age").cast("double")) \
  .withColumn("joining_date", col("joining_date").cast("date")) \
  .printSchema()
```
